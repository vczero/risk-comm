## 标题语气和点击量（阅读量）关系

```python
# The tone of the title
def tone_em_data_clean():
    # rows = []
    # rs = []
    # for i, r in df_em.iterrows():
    #     em_label = r['em_label'].replace('分', '')
    #     em_label = em_label.replace('。', '')
    #     em_label = em_label.replace(' ', '')
    #     rows.append([r['link'], em_label])
    #     rs.append(em_label)
    
    # print('语气紧急程度：', list(set(rs)))
    # temp = pd.DataFrame(rows, columns=['link', 'em_label'])
    temp = df_em.drop_duplicates(subset='link')
    return temp


def show_tone_em_read_num():
    t1 = df_typhoon[['link', 'read_num']]
    t2 = tone_em_data_clean()
    t3 = pd.merge(t1, t2, on='link')
    t4 = t3[['read_num', 'em_label']]
    t4.to_csv('./R_Vis/factor_tone.csv', index=None)
    return t4

df_tone = show_tone_em_read_num()
df_tone
```

```python
def show_tone_mean_median():
    NUM = 100000
    total_10w = df_tone[df_tone.read_num >= NUM]
    print('10w+文章数量：', len(total_10w))
    types = [3, 2, 1]
    for i in types:
        temp = df_tone[df_tone.em_label==i]
        temp_10w = temp[temp.read_num >=NUM]
        print('===============================')
        print('type', i)
        print('文章总数：', len(temp))
        print('10w+文章数量：', len(temp_10w))
        print('10w+文章占比：', len(temp_10w) / len(temp) * 100)
        print('不是10w+文章数量：', len(temp) - len(temp_10w))
        print(temp.describe())
show_tone_mean_median()
```

## 标题长度和点击量（阅读量）关系

```python
# 使用R可视化
df_typhoon[['read_num', 'title_len']].to_csv('./R_Vis/factor_title_len.csv', index=None)
```

```python
def show_title_len_group():
    NUM = 100000
    temp = df_typhoon[['read_num', 'title_len']].copy()
    bins = [1, 16, 32, 48, 64]  
    labels = ['1-16', '17-32', '33-48', '49-64']
    temp['group'] = pd.cut(temp['title_len'], bins=bins, labels=labels, right=True)

    grouped = temp.groupby('group', observed=False)
    for group_name, group_data in grouped:
        t_10w = group_data[group_data.read_num >=NUM]
        
        print('==================================================')
        print(f"Group: {group_name}")
        print('文章总数：', len(group_data))
        print('10w+文章数量：', len(t_10w))
        print('10w+文章占比：', len(t_10w) / len(group_data) * 100)
        print('不是10w+文章数量：', len(group_data) - len(t_10w))
        
        print(group_data['read_num'].describe())
        
show_title_len_group()
```

## 符号和阅读量的关系

```python
# 比较感叹号
def show_E(df, NUM):
    df_yes = df[df.is_exc_point==1]
    df_no = df[df.is_exc_point==0]

    print('感叹号数量：', len(df_yes))
    print(df_yes['read_num'].describe())
    print('=================================')
    print('无感叹号数量：', len(df_no))
    print(df_no['read_num'].describe())
    # read_10w_YES = df_yes[df_yes.read_num >= NUM]
    # read_10w_No = df_no[df_no.read_num >= NUM]
    
    # print('包含感叹号10w+的数量：', len(read_10w_YES))
    # print('包含感叹号10w+的占比：', len(read_10w_YES)/len(df_yes))
    # print('不包含感叹号10w+的数量：', len(read_10w_No))
    # print('不包含感叹号10w+的占比：', len(read_10w_YES)/len(read_10w_No))

def contain_exclamation():
    NUM = 100000
    rows = []
    for i, r in df_typhoon[['link', 'title', 'read_num']].iterrows():
        is_excl = 0
        if '！' in r['title'] or '!' in r['title']:
            is_excl = 1
        rows.append([r['read_num'], is_excl])
    temp = pd.DataFrame(rows, columns=['read_num', 'is_exc_point'])

    show_E(temp, NUM)
    
    temp.to_csv('./R_Vis/factor_exclamation.csv', index=None)
    return temp

contain_exclamation()
```

```python
# 问号
def show_Q(df, NUM):
    df_yes = df[df.is_question_mark==1]
    df_no = df[df.is_question_mark==0]

    print('问号数量：', len(df_yes))
    print(df_yes['read_num'].describe())
    print('=================================')
    print('无问号数量：', len(df_no))
    print(df_no['read_num'].describe())
    
def contain_question():
    NUM = 100000
    rows = []
    for i, r in df_typhoon[['link', 'title', 'read_num']].iterrows():
        is_Q = 0
        if '?' in r['title'] or '？' in r['title']:
            is_Q = 1
        rows.append([r['read_num'],is_Q])
    temp = pd.DataFrame(rows, columns=['read_num','is_question_mark'])

    show_Q(temp, NUM)
    temp.to_csv('./R_Vis/factor_question.csv', index=None)
    return temp

contain_question()
```
